```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# 1. Setup the data
n_snapshots = 500
n_atoms = 50
raw_data = np.random.rand(n_snapshots, n_atoms * 3) 

# 2. Pre-process (Center the data)
mean_centered_data = raw_data - np.mean(raw_data, axis=0)

# 3. Calculate PCA (This defines 'reduced_data')
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(mean_centered_data)

# 4. Visualize the results
plt.figure(figsize=(10, 7))

# Use time_steps to color the dots so the 'path' of the protein can be seen
time_steps = np.arange(n_snapshots)

plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=time_steps, cmap='viridis', alpha=0.7)
plt.colorbar(label='Snapshot Index (Time progression)')
plt.title('Protein Conformational Landscape: Detection of Structural Patterns')
plt.xlabel('Principal Component 1 (Major Transition)')
plt.ylabel('Principal Component 2 (Minor Transition)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
